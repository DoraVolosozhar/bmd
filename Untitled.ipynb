```python
import matplotlib.pyplot as mp
```

```python
%matplotlib inline 
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2 

x
```

```python
y
```

```python
mp.plot(x,y)
mp.xlabel('X value')
mp.ylabel('Y value') 
mp.title('Title')
```

```python
mp.subplot(1,2,1)
mp.plot(x,y,'r')

mp.subplot(1,2,2)
mp.plot(x,y,'b')
```

```python
gr=mp.figure() 
axes=gr.add_axes([0,0,1,1]) 

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
gr=mp.figure() 

axes1=gr.add_axes([0,0,1,1]) 
axes2=gr.add_axes([0.1,0.5,0.4,0.5])

axes1.plot(x,y,'r')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Title') 

axes2.plot(x,y,'b')
axes2.set_xlabel('X label')
axes2.set_ylabel('Y label')
axes2.set_title('Title')
```

```python
fig,axes=mp.subplots()
axes.plot(x,y,'r')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title') 
```

```python
fig,axes=mp.subplots(nrows=1,ncols=2) 
axes[0].plot(x,y) 
axes[1].plot(y,x)
```

```python
fig,axes = mp.subplots(nrows=4, ncols=4)
mp.tight_layout()
```

```python
fig=mp.figure()
ax=fig.add_axes([0,0,1,1]) 
ax.plot(x,x**2, label="x**2")
ax.plot(x,x**3, label="x***3")
ax.legend()
```

```python
fig=mp.figure(figsize=(1,5),dpi=100) #dpi - количество точек на дюйм 
ax=fig.add_axes([0,0,1,1]) 
ax.plot(x,x**2, label="x**2")
ax.plot(x,x**3, label="x***3")
ax.legend()
```

```python
fig=mp.figure() 
ax=fig.add_axes([0,0,1,1]) 
ax.plot(x,x**2, label="x**2",color='b',lw=2,ls='--',marker="v",markersize=5,markerfacecolor='y')
ax.plot(x,x**3, label="x***3",color='r',lw=5,ls='--')
ax.legend()
```
